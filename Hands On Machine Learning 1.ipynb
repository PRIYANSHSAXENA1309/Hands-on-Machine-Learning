## Linear Regression, Ridge and Lasso

```python
from sklearn.datasets import load_boston
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
%matplotlib inline
```

```python
df=load_boston()
```

```python
df
```

```python
dataset=pd.DataFrame(df.data)
dataset.columns=df.feature_names
dataset.head()
```

```python
dataset['Price']=df.target
```

```python
dataset.head()
```

```python
X=dataset.iloc[:,:-1]
Y=dataset.iloc[:,-1]
```

```python
print(X.head())
print(Y.head())
```

```python
## linear Rigression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
lin_reg=LinearRegression()
nmse=cross_val_score(lin_reg,X,Y,scoring='neg_mean_squared_error',cv=5)
mean=np.mean(nmse)
print(mean)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.4,random_state=1)
```

```python
lin_reg.fit(xtrain,ytrain)
pred=lin_reg.predict(xtest)
```

```python
from sklearn.metrics import r2_score
score=r2_score(ytest,pred)
print(score)
```

```python
## Ridge Regression
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge=Ridge()
parameter={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}
ridge_regressor=GridSearchCV(ridge,parameter,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X,Y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
## Lasso Regression
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso=Lasso()
parameter={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
lasso_regressor=GridSearchCV(lasso,parameter,scoring='neg_mean_squared_error',cv=10)
lasso_regressor.fit(X,Y)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
```

```python
df1=load_breast_cancer()
dataset1=pd.DataFrame(df1.data)
dataset1.columns=df1.feature_names
dataset1.head()
```

```python
dataset1['Target']=df1.target
dataset1.head()
```

```python
X1=dataset1.iloc[:,:-1]
Y1=dataset1.iloc[:,-1]
```

```python
print(X1)
print(Y1)
```

```python
dataset1['Target'].value_counts()
```

```python
xtrain1,xtest1,ytrain1,ytest1=train_test_split(X1,Y1,test_size=0.33,random_state=42)
```

```python
parameter1=[{'C':[1,5,10]},{'max_iter':[100,150]}]
```

```python
model1=LogisticRegression(C=100,max_iter=100)
model=GridSearchCV(model1,param_grid=parameter1,scoring='f1',cv=5)
```

```python
model.fit(xtrain1,ytrain1)
```

```python
print(model.best_params_)
print(model.best_score_)
```

```python
pred1=model.predict(xtest1)
pred1
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix(ytest1,pred1)
```

```python
accuracy_score(ytest1,pred1)
```

```python
print(classification_report(ytest1,pred1))
```
